```python
from pathlib import Path
import torch
from diffusers import StableDiffusionPipeline, EulerAncestralDiscreteScheduler

LAB_ROOT = Path(r"D:\Learning\Year 3\Semester 6\Computer Vision\Computer-Vision-ITMO-2026\Lab 8")

LORA_PATH = LAB_ROOT / "skscuong_lora.safetensors"
OUTPUT_DIR = LAB_ROOT / "outputs"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("LoRA exists:", LORA_PATH.exists())
print("CUDA:", torch.cuda.is_available())
print("Device:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")
```

```python
base_model = "runwayml/stable-diffusion-v1-5"

dtype = torch.float16 if torch.cuda.is_available() else torch.float32
device = "cuda" if torch.cuda.is_available() else "cpu"

pipe = StableDiffusionPipeline.from_pretrained(
    base_model,
    torch_dtype=dtype,
    safety_checker=None
)

pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to(device)

pipe.load_lora_weights(str(LORA_PATH))

print("Pipeline loaded.")
```

```python
prompts = [
    "photo of skscuong man in a forest, high quality, realism",
    "photo of skscuong man in a city, high quality, realism",
    "photo of skscuong man in a beach, high quality, realism",
    "photo of skscuong man in cyberpunk style, neon lights, high quality, realism",
    "photo of skscuong man made of metal, high quality, realism",
    "photo of skscuong man as a futuristic astronaut, high quality, realism",
]

for i, prompt in enumerate(prompts, start=1):
    image = pipe(
        prompt=prompt,
        negative_prompt=None,
        num_inference_steps=28,
        guidance_scale=7.5,
        height=512,
        width=512
    ).images[0]

    save_path = OUTPUT_DIR / f"lab8_{i}.png"
    image.save(save_path)
    print("Saved:", save_path)
```
