### Task 1: NumPy random array + operations

```python
import numpy as np

# make every time to generate the same result
np.random.seed(42)

# 1) Generate random array
A = np.random.randint(0, 100, size=(4, 5))   #number from 0 -> 100, size 4x5
B = np.random.randint(0, 100, size=(4, 5))

print("A:\n", A)
print("B:\n", B)
```

```python
# 2) Operations with array

# Operations between A and B
print("\n[Arithmetic operations]")
print("A + B:\n", A + B)
print("A - B:\n", A - B)
print("A * B (Hadamard):\n", A * B)
print("A / B:\n", A / B)

# Operations between A and numerics
print("A + 10:\n", A + 10)
print("A ** 2 : \n", A * 2)
print("A / 2:\n", A / 2)
print("A - 15:\n", A - 15)
```

```python
# Statistical functions

print("\n[Statistical functions]")
print("\n")
print("- Sum:", A.sum())
print("- Mean:", A.mean())
print("- Min:", A.min())
print("- Max:", A.max())
print("- Std:", A.std())
print("- Sum by rows:", A.sum(axis=1))
print("- Mean by columns:", A.mean(axis=0))
```

```python
# Indexing and Slicing

print("\n[Indexing & slicing]")
print("\n")
print("- Element A[1,2]:", A[1,2])
print("- First row:", A[0,:])
print("- Second column:", A[:,1])
print("- Submatrix A[0:2,1:3]:\n", A[0:2,1:3])
```

```python
print("\n[Boolean masking]")
mask = A > 50
print("Mask (A>5):\n", mask)
print("Filtered values:", A[mask])
```

```python
print("\n[5) Reshape]")

print("Original shape:", A.shape)

A_2x10 = A.reshape(2, 10)
print("\n A reshaped to (2,10):\n", A_2x10)

A_5x4 = A.reshape(5, 4)
print("\n A reshaped to (5,4):\n", A_5x4)

A_flat = A.reshape(-1)  # flatten về vector 1D
print("\n A flattened shape:", A_flat.shape)
print("A_flat:\n", A_flat)
```

```python
print("\n[6) Broadcasting]")

v = np.array([1, 2, 3, 4, 5])  # shape (5,)
print("v shape:", v.shape)

A_plus_v = A + v
print("A + v:\n", A_plus_v)
```

```python
print("\n[7) Dot product]")

dot_row = np.dot(A[0], B[0])  # (5,) dot (5,)
print("A[0] · B[0] =", dot_row)
```

### Task 2: Object Detection on image with OpenCV

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("image.jpg")
if img is None:
    raise FileNotFoundError("image not found")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# little blur to decrease noises
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# canny edges
edges = cv2.Canny(blurred, 50, 150)

# find ALL contours
contours, hierarchy = cv2.findContours(
    edges.copy(),
    cv2.RETR_EXTERNAL,
    cv2.CHAIN_APPROX_SIMPLE
)

print(f"number contours: {len(contours)}")

# draw
img_all_contours = img_rgb.copy()
cv2.drawContours(img_all_contours, contours, -1, (0, 255, 0), 2)



# show off
plt.figure(figsize=(16,4))

plt.subplot(1,4,1); plt.title("Original"); plt.imshow(img_rgb); plt.axis("off")
plt.subplot(1,4,2); plt.title("Gray"); plt.imshow(gray, cmap="gray"); plt.axis("off")
plt.subplot(1,4,3); plt.title("Edges (Canny)"); plt.imshow(edges, cmap="gray"); plt.axis("off")
plt.subplot(1,4,4); plt.title("ALL contours"); plt.imshow(img_all_contours); plt.axis("off")

plt.tight_layout()
plt.show()
```

### Task 3 : Using NumPy slicing to draw a rectangle onto an image

```python
def rectangle_np(img, pt1, pt2, color=(0, 255, 0), thickness=2):

    out = img.copy()
    H, W = out.shape[:2]

    x1, y1 = pt1
    x2, y2 = pt2

    # sort
    x1, x2 = sorted([int(x1), int(x2)])
    y1, y2 = sorted([int(y1), int(y2)])

    # clamp to edges of image
    x1 = max(0, min(x1, W - 1))
    x2 = max(0, min(x2, W - 1))
    y1 = max(0, min(y1, H - 1))
    y2 = max(0, min(y2, H - 1))

    # if thickness minus => fill all
    if thickness is not None and thickness < 0:
        out[y1:y2+1, x1:x2+1] = color
        return out

    t = max(1, int(thickness))

    # top border
    out[y1:y1+t, x1:x2+1] = color
    # bottom border
    out[y2-t+1:y2+1, x1:x2+1] = color
    # left border
    out[y1:y2+1, x1:x1+t] = color
    # right border
    out[y1:y2+1, x2-t+1:x2+1] = color

    return out
```

```python
# draw on image.jpg

img_bgr = cv2.imread("image.jpg")
if img_bgr is None:
    raise FileNotFoundError("not found image.jpg")

demo1 = rectangle_np(img_bgr, (50, 50), (300, 220), color=(0, 255, 0), thickness=4)

# Convert to RGB
img_rgb   = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
demo1_rgb = cv2.cvtColor(demo1, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1); plt.imshow(img_rgb);   plt.title("Original"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(demo1_rgb); plt.title("rectangle_np result"); plt.axis("off")
plt.show()
```

```python
# draw on blank image

blank = np.zeros((400, 600, 3), dtype=np.uint8)
demo2 = rectangle_np(blank, (60, 60), (540, 320), color=(255, 255, 255), thickness=6)  # viền trắng
demo2 = rectangle_np(demo2, (120, 120), (300, 260), color=(0, 255, 255), thickness=3) # tô kín vàng (BGR)

plt.figure(figsize=(12,4))
plt.imshow(cv2.cvtColor(demo2, cv2.COLOR_BGR2RGB))
plt.title("rectangle_np on blank image")
plt.axis("off")
plt.show()
```
